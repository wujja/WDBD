```python
file = None
with open('opowiesc.txt', 'r') as f:
    file = f.read().split()
file = [''.join(filter(str.isalnum, x)).lower() for x in file] 
file
```

```python
from multiprocessing import Pool
from functools import reduce
import time

start = time.process_time()

def count_words(words):
    
    dict_count = {}
    for word in words:
        if(word in dict_count):
            dict_count[f'{word}'] += 1
        else:
            dict_count[f'{word}'] =  1

    return dict_count

mapper = count_words

def reducer(a, b):
    final_dict = a
    for word in b:
        if(word in final_dict.keys()):
            final_dict[f'{word}'] += b[word]
        else:
            final_dict[f'{word}'] =  b[word]
    return final_dict


def chunkity(dane, liczba_chunks):
    return(dane[i::liczba_chunks] for i in range(liczba_chunks))


data_chuck = chunkity(file, 4)
pool = Pool(4)

mapped = pool.map(count_words, data_chuck)
#print(mapped)

reduced = reduce(reducer, mapped)
print(reduced)
print(time.process_time() - start)

```
