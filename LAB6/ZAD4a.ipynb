```python
from multiprocessing import Pool
from functools import reduce
import random

def random_list(n):
    random_list =[]
    for i in range(0,n):
        number = random.randint(1, 2234523)
        random_list.append(number)
    return random_list

list_list = random_list(20)

def reducer(a, b):
    if(a > b):
        return a
    return b

def chunkity(dane, liczba_chunks):
    return(dane[i::liczba_chunks] for i in range(liczba_chunks))


def chunks_mapped(chunk):
    return reduce(reducer,chunk)

data_chuck = chunkity(list_list, 4)

pool = Pool(4)

mapped = pool.map(chunks_mapped, data_chuck)


reduced = reduce(reducer, mapped)

print(reduced)
```
